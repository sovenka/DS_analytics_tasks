Тренировочные данные не являются непрерывными. Ест "пропуски" во временной шкале с данными. 

```python
#Посмотрим, с каким шагом фиксируется время и каковы самые большие пропуски
timedelta = []
for i in range(len(train_data)-1):
    timedelta.append(train_data.loc[i+1, ['date_upd']] - train_data.loc[i, ['date_upd']])
    
timedelta = pd.DataFrame(timedelta)
timedelta.describe()
```

```python
timedelta.head()
#В основном, данные пишутся каждый час.
```

```python
timedelta.query('date_upd != "0 days 01:00:00"').describe()
```

```python
timedelta.query('date_upd == "122 days 01:00:00"')
```

Данные у нас с 15 января 2016 года по 18 августа 2018. Основная дискретность 1 час, однако есть и пропуски в 2% случаев (то есть, данные идут не подряд). Медианный пропуск 3 часа, максимальный 122 дня. Большие пропуски (4 месяца) соответствуют осени-зиме 2016 и осени-зиме 2017 года.

### Создание функции для вычисления симметричной среднего абсолютного отклонения sMAPE 

```python
#Зададим функцию для Smape_rough(clean):
def sMAPE(predicted_values, target_values):
    n = len(predicted_values)
    chislitel = np.abs(predicted_values - target_values)
    znamenatel = (np.abs(predicted_values) + np.abs(target_values))*0.5
    smape = np.sum(chislitel/znamenatel/n)
    return smape[0]*0.25 + smape[1]*0.75

```
